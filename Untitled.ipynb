```python
def start_at_same_time(dataset):
    
    headers = dataset.keys()
    val_of_interest = dataset[headers[1]]
    dates = dataset["DATE"]
    
    dataset_after_1986 = [] 
    after_1986 = False
    
    for i in range(len(dates)):
        if "1987" in dates[i]:
            after_1986 = True
        if after_1986:
            dataset_after_1986.append(val_of_interest[i])

    np_dataset_after_1986 = np.array(dataset_after_1986)
    

    
    output_array = []
    
    month = dates[0].split('-')[1]
    
    j = 0
    for i in range(len(np_dataset_after_1986)):
        new_month = dates[i].split('-')[1]
        
        if month != new_month:
            month = new_month
            output_array.append(val_of_interest[i])
            
        
        date = dates[i]
    
    return output_array
```

### Now we will test the null hypothesis using the F statistic. The null, H_{0}, is that beta_{0} = beta_{1} = ... = beta_{n} = 0. In this case we will use a pretty standard alpha value of 0.05. If our p-value is less than alpha, or if our observed F statistic is greater than the one that we should observe, then we will reject the null hypothesis and determine that there is some correlation between these variables. We'll make an ANOVA table:



#### df                  ss                                       ms                      F

##### p-1        sum(y_pre_i - y_avg)**2            SSR/(p-1)                 mse/msr   

##### n-p        sum(y_i - y_hat_i)**2                 SSE/(n-p)

##### n-1

```python
my_inputs = [M2_monthly_test, 
             MB_monthly_test, 
             ratioMBM2_monthly_test,
             Loans_monthly_test,
             case_shiller_vals_test]

my_output = diff_test

my_coefs = model_inqeuality_train.coef_

my_intercept = model_inqeuality_train.intercept_


def calculate_f(inputs, output, coefficients, intercept):
    p = len(inputs) + 1
    n = len(output)

    #assert (len(inputs) == 1) or (len(inputs) == len(coefficients))
    errs_sq = []
    regs_sq = []
    output_mean = sum(output)/len(output)

    for i in range(len(inputs[0])):
        
        output_hat = intercept
        for j in range(len(inputs)):
            output_hat += inputs[j][i]*coefficients[j]
            output_actual = output[i]
            
        err = output_actual - output_hat
        errs_sq.append(err**2)
        regr = output_hat - output_mean
        regs_sq.append(regr**2)
        
    SSE_ = sum(errs_sq)
    MSE_ = (SSE_/(n-p))
    
    SSR_ = sum(regs_sq)
    SST_ = SSE_ + SSR_
    fraf = (n-1)/(n-p)
    R_sq = round(1- (fraf)*(SSE_/SST_), 3)
    print("R_sq = ", R_sq)
    MSR_ = (SSR_/(p-1))
    
    F = round(MSR_/MSE_, 2)
    print("p = ",p,"n = ",n,"(top entry:", p-1,"side entry:",n-p,")")
    return F





def do_all_fs(inputs, output, coefficients, intercept):
    print("0: M2")
    print("1: Monetary Base")
    print("2: Ratio of Monetary Base to M2")
    print("3: Total United States Housing Loans")
    print("4: Case-Shiller Index")
    print()
    print()
    print()
    x = [[0], [1], [2], [3], [4], 
         [0,1], [0,2], [0,3], [0,4],
         [1,2], [1,3], [1,4], 
         [2,3], [2,4],
         [3,4],
         [0,1,2], [0,1,3], [0,1,4], [0,2,3], [0,2,4], [0,3,4],
         [1,2,3], [1,2,4], [1,3,4],
         [2,3,4],
         [0,1,2,3], [0,1,2,4], [0,1,3,4], [0,2,3,4], [1,2,3,4],
         [0,1,2,3,4]]
    for i in range(len(x)):
        print(str(x[i])+":")
        the_inputs = []
        the_coefs = []
        for j in range(len(x[i])):
            the_inputs.append(inputs[x[i][j]])
            the_coefs.append(coefficients[x[i][j]])
            
            
        f_stat = calculate_f(the_inputs, output, the_coefs, intercept)
        print("F STATISTIC:",f_stat)
        print()
        print()
        print()
        print()
do_all_fs(my_inputs, my_output, my_coefs, my_intercept)
#calculate_f(my_inputs, my_output, my_coefs, my_intercept)
```
